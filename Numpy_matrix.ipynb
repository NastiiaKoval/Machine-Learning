Завдання 1

Створіть одновимірний масив (вектор) з числами Фібоначчі, де кожне число
дорівнює сумі двох попередніх, та виведіть його значення.


```python
# Функція для генерації чисел Фібоначчі
def fibonacci(n):
    fib_sequence = [0, 1]
    for i in range(2, n):
        fib_sequence.append(fib_sequence[-1] + fib_sequence[-2])
    return np.array(fib_sequence)

# Створення одновимірного масиву з першими 10 числами Фібоначчі
fib_array = fibonacci(10)

# Виведення масиву
print(fib_array)
```

Завдання 4

 Створіть масив розміром 3x3, заповніть його випадковими дійсними числами від 0 до 1 та знайдіть середнє геометричне всіх елементів.

```python
# Створення масиву 3x3 з випадковими дійсними числами від 0 до 1
array = np.random.uniform(0, 1, (3, 3))

# Обчислення середнього геометричного через numpy
geometric_mean = np.exp(np.mean(np.log(array)))

# Виведення масиву та середнього геометричного
print("Масив:\n", array)
print("Середнє геометричне значення:", geometric_mean)
```

Завдання 5

Створіть два одновимірних масиви розміром 6, заповніть їх випадковими дійсними
числами від -5 до 5 та виконайте на них поелементні операції додавання, віднімання
та множення

```python
# Створення двох одновимірних масивів з випадковими дійсними числами від -5 до 5
array1 = np.random.uniform(-5, 5, 6)
array2 = np.random.uniform(-5, 5, 6)

# Поелементне додавання
sum_arrays = array1 + array2

# Поелементне віднімання
diff_arrays = array1 - array2

# Поелементне множення
prod_arrays = array1 * array2

# Виведення результатів
print("Перший масив:", array1)
print("Другий масив:", array2)
print("Додавання:", sum_arrays)
print("Віднімання:", diff_arrays)
print("Множення:", prod_arrays)
```

Завдання 8

Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 15 та знайдіть її обернену матрицю.

```python
matrix = np.random.randint(1, 16, size=(2, 2))

# Перевірка, чи матриця є оберненою (детермінант не дорівнює нулю)
det = np.linalg.det(matrix)
if det != 0:
    # Знаходження оберненої матриці
    inverse_matrix = np.linalg.inv(matrix)
    print("Матриця:\n", matrix)
    print("\nОбернена матриця:\n", inverse_matrix)
else:
    print("Матриця не має оберненої (детермінант дорівнює нулю):\n", matrix)
```

Завдання 15

Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть вектор-стовпчик, що містить суму квадратів елементів кожного рядка матриці.

```python
# Створення матриці 3x3 з випадковими дійсними числами від 0 до 1
matrix = np.random.uniform(0, 1, size=(3, 3))

# Обчислення суми квадратів елементів кожного рядка
row_sums_of_squares = np.sum(matrix**2, axis=1).reshape(-1, 1)

# Виведення матриці та вектора-стовпчика з сумами квадратів
print("Матриця:\n", matrix)
print("\nВектор-стовпчик сум квадратів елементів кожного рядка:\n", row_sums_of_squares)
```
